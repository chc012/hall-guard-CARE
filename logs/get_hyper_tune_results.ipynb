```python
import pandas as pd
```

```python
file_name = "sbatch-job-7074535-edited.out"
```

```python
readlines = open(file_name, encoding="utf-8").readlines()
readlines = [line.strip().split() for line in readlines if line[0] == "[" and "pruned" not in line][1:]

d = {
    "Accuracy": [float(l[8]) for l in readlines],
    "Learning Rate": [float(l[12][:-1]) for l in readlines],
    "Batch Size": [int(l[14][:-2]) for l in readlines]
}

df = pd.DataFrame(d)
df = df.sort_values(by=["Batch Size", "Learning Rate"], ascending=True)
df.to_latex("learning_rate_batch_size.tex", index=False, formatters=["{:.4f}".format, "{:.2e}".format, "{}".format])
df
```
